```python
def get_w (x, y):
    ones = np.ones(x.shape[0])
    x = np.column_stack([ones, x])
    xTx = x.T.dot(x)
    xTx_inv = np.linalg.inv(xTx)
    w = xTx_inv.dot(x.T).dot(y)
    return w[0], w[1:]
```

```python
base = ['engine_displacement', 'horsepower', 'vehicle_weight']
```

```python
df_train[base].isnull().sum()
```

```python
def prep_x(df):
    df_num = df[base]
    df_num = df_num.fillna(0)
    x = df_num.values
    return x
```

```python
mean_eff = (df_train.horsepower.mean())
mean_eff
```

```python
def prep_x1(df, m):
    df_num = df[base]
    df_num = df_num.fillna(m)
    x = df_num.values
    return x
```

```python
def rmse(y, y_pred):
    se = (y-y_pred)**2
    mse = se.mean()
    return np.sqrt(mse)
```

```python
x_train = prep_x(df_train)
w0, w = get_w(x_train, y_train)

x_val = prep_x(df_val)
y_pred = w0 + x_val.dot(w)
e = rmse(y_val, y_pred)
round(e, 2)

```

```python
x_train = prep_x1(df_train, mean_eff)
w0, w = get_w(x_train, y_train)

x_val = prep_x1(df_val, mean_eff)
y_pred = w0 + x_val.dot(w)
e = rmse(y_val, y_pred)
round(e, 2)
```
